```python
#para susbstituents seq: parent, Bromo, CF3, CH3, Chloro, Fluoro, Nitro, OMe, isopropyl, NMe2; energy is in kcal/mol
Hamment_const_list_para=[0,0.23,0.54,-0.17,0.23,0.56,0.78,-0.27,-0.15,-0.83,-0.01]
Free_energy_list_para=[4.40756,5.546472500620894,8.994584999976496,1.997332500022253,5.064552499840147,3.862262500022098,11.404812500097705,-1.7500975000757535,1.9170125000823646,-10.31170750000399,3.0452574999446824]
```

```python
linear_model_para=np.polyfit(Hamment_const_list_para,Free_energy_list_para,1)
linear_model_para_fn=np.poly1d(linear_model_para)
x_s=np.arange(-1,2)
plt.figure(figsize=(6,6))
plt.plot(Hamment_const_list_para,Free_energy_list_para,marker='x',linestyle='')
plt.plot(x_s,linear_model_para_fn(x_s),color="green")
plt.xlabel('σ(para)')
plt.ylabel('△G(kcal/mol)')
plt.show()
result = scipy.stats.linregress(Hamment_const_list_para, Free_energy_list_para)
print('The Pearson correlation coefficient is',result.rvalue)
```

```python
#meta substituent seq: parent, CH3, Bromo, isopropyl; energy is in kcal/mol
Hamment_const_list_meta=[0,-0.07,0.39,-0.04]
Free_energy_list_meta=[4.40756,2.632362499988403,7.163540000176454,3.356497500014939]
```

```python
linear_model_meta=np.polyfit(Hamment_const_list_meta[:-1],Free_energy_list_meta[:-1],1)
linear_model_meta_fn=np.poly1d(linear_model_meta)
x_s_meta=np.arange(-0.5,1)
plt.figure(figsize=(6,6))
plt.plot(Hamment_const_list_meta,Free_energy_list_meta,marker='x',linestyle='')
plt.plot(x_s_meta,linear_model_meta_fn(x_s_meta),color="green")
plt.xlabel('σ(meta)')
plt.ylabel('△G(kcal/mol)')
result2 = scipy.stats.linregress(Hamment_const_list_meta, Free_energy_list_meta)
plt.show()
print('The Pearson correlation coefficient is',result2.rvalue)
```
